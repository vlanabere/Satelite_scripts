## Global Precipitation Measurement (GPM) Microwave Imager (GMI)

![title](http://www.smithsconnectors.com/getfile/cd9f7043-f588-4487-8324-c92f802b0f29/sub-banner.jpg.aspx)

GMI es un radiómetro pasivo de microondas con un escaneo de tipo cónico a bordo del satélite GPM. Este satélite orbita la Tierra a una altura de 407 kilómetros en una órbita circular con una inclinación de 65° y un ancho de pasada (*swath*) de 885 kilómetros.

![title](GPM_GMI_scan_geometry.jpg)

El *swath S1* esta conformado por 9 (nueve) canales similares a los del radiómetro TMI que estaba a bordo del satélite TRMM (10.65V, 10.65H, 18.7V, 18.7H, 23.8V, 36.5V, 36.57H, 89V y 89H). Mientras que el *swath S2* lo conforma 4 (cuatro) canales posicionados en frecuencias similares al radiómetro AMSU-B a bordo de los satélites de la serie NOAA (165.5V, 165.5H, 183.31+/-3V, and 183.31+/-7V). 

- Los canales de frecuencias bajas (S1) son para medir las señales provenientes de la emisión por parte de la superficie y la atmósfera.


- Los canales de frecuencia alta (S2) miden la disminución de la temperatura de brillo por efecto de la dispersión por parte de las partículas de hielo. 

En la siguiente tabla se resumen las principales características de cada uno de los canales que componen el sensor. 

![title](GPM_GMI_channel_table.png)

#### Relación entre los *swath  S1* y *S2*
*Swath S2* posee el mismo número de escaneos y el mismo número de píxeles que el *swath S1*. Cada escaneo S1 contiene nueve canales muestreados 221 veces a lo largo del escaneo. Cada escaneo S2 contiene cuatro canales muestreados 221 veces a lo largo de cada escaneo. Mientras que el ángulo de incidencia del *swath S1* es diferente al del *swath S2*, la ubicación geográfica del centro de cada pixel es diferente (aprox. 150 km en la dirección de vuelo del satélite).

### 2.1 Comentarios generales

- La cantidad de radiación de microondas emitida por la superficie terrestre depende de las interacciones entre la energía y los diversos elementos y estructuras que componen la superficie. 
- Las dos propiedades que afectan en medida considerable la cantidad de radiación de microondas emitida son la **polarización** y el **efecto dieléctrico**. Cada propiedad varía según la longitud de onda y las características físicas del material reflector o emisor. Esto permite distinguir los elementos sólidos de los líquidos y congelados, tanto sobre tierra firme como en las superficies oceánicas.
- En regiones abiertas con vegetación relativamente escasa, el contenido de humedad del suelo superficial es el factor predominante en las emisiones de radiación de microondas desde la superficie. 
- Una de las propiedades electromagnéticas más importantes para la observación de las superficies en la región de microondas es el efecto dieléctrico. 
- Dicho efecto explica la mayor parte de la reflexión y dispersión que ocurre a medida que la radiación interactúa con las moléculas de la superficie y suele cuantificarse mediante un término conocido como la constante dieléctrica. 
- La constante dieléctrica del agua es mucho mayor que la del suelo seco y existe una fuerte correlación entre el efecto dieléctrico y el reflejo (y, por ende, la emisividad). 
- La emisividad es inversamente proporcional al efecto dieléctrico, de modo que a medida que el efecto dieléctrico (reflejo) aumenta, la emisividad (o eficiencia de emisión) disminuye.
- La introducción de agua en el suelo produce un aumento drástico en la constante dieléctrica y una correspondiente disminución en la emisividad del suelo.

### 2.2 Curvas de emisión
Esta gráfica muestra los valores de emisividad en el rango de las microondas para varios tipos de superficie. ### 2.3 Profundidad de penetración en el suelo en función de la frecuencia
- Otra consideración importante a la hora de observar el suelo con longitudes de onda de microondas es la ***profundidad de penetración***.
- Se define como la ***capa superior del suelo en la cual se origina el 63% de la radiación terrestre emitida a una frecuencia o longitud de onda dada***.
- La profundidad de penetración es mayor para las frecuencias de microondas más bajas (1 a 10 GHz, por ejemplo), porque el suelo absorbe y dispersa las longitudes de onda mayores en menor medida. 
- En contraste, la penetración es considerablemente menor a frecuencias más altas (de 85 a 89 GHz, por ejemplo) y estos canales sólo comunican información sobre la capa superior del suelo de unos milímetros de espesor.
- Una buena regla general para las observaciones de suelos relativamente secos con sistemas de microondas pasivos es que la profundidad de penetración equivale aproximadamente a la longitud de onda de la radiación. La siguiente tabla muestra las profundidades de penetración aproximadas para las frecuencias de uso común en varios sensores de microondas actuales.

![table](sm_pd_table.gif)

- Otro factor que influye y limita la profundidad de penetración de microondas en el suelo es el contenido de humedad del suelo.
- Cuando aumenta el contenido de humedad del suelo, la profundidad de penetración disminuye.
- Recordar que una capa de suelo relativamente mojada dispersa y refleja más energía, de modo que su emisividad es menor en comparación con un suelo seco.
- Este mayor grado de dispersión y reflexión bloquea parte de la radiación e impide que alcance la superficie, de modo que un satélite percibe progresivamente menos energía de capas cada vez más profundas.

### 2.4 Factores a tener en cuenta
- La vegetación y la rugosidad de la superficie interfieren con la señal emitida por la superficie terrestre.
- La vegetación emite su propia radiación de microondas, que puede confundirse con la emisión del suelo que se desea capturar.
- La vegetación aumenta la emisividad de una superficie que, en realidad, está mojada. Esto aumenta la temperatura de brillo de microondas observada y da a la superficie una apariencia más seca de lo que es.

![title](sm_vegetation_scgrab2.gif)

- Afortunadamente, la rugosidad de la superficie y la vegetación producen efectos de polarización muy específicos en la señal de microondas.
- Se pueden comparar varias frecuencias y las polarizaciones vertical y horizontal para distinguir los efectos de la vegetación y la rugosidad de la superficie de los de la humedad de la superficie.
- A medida que una superficie se humedece, el grado de polarización se vuelve más pronunciado, especialmente a frecuencias más bajas.

![title](surface_emissivity_spectra_Vpol.gif)

![title](surface_emissivity_spectra_Hpol.gif)

### 2.5 Húmedad en el suelo: índice normalizado de polarización
La técnica empleada para generar estimaciones de la humedad del suelo depende de un parámetro conocido como **razón de polarización normalizada** o **índice normalizado de polarización**, que se calcula dividiendo la diferencia en la temperatura de brillo de los canales de polarización vertical y horizontal y su suma correspondientes a las frecuencias de 10.7 GHz y 18.7 GHz para el caso del radiómetro AMSR-E a bordo de los satélites Aqua y Terra. 

![title](polratio1_243.gif)

A medida que la humedad del suelo aumenta, la diferencia en la polarización (numerador) aumenta y la razón se vuelve más grande.

![title](polratio2_243.gif)

Los valores de humedad del suelo se calculan en términos del desvío de la razón de polarización de 10.7 GHz desde un valor basal mínimo que se calcula en cada punto de una retícula para un ciclo anual. También se utiliza el canal de 18.7 GHz, el cual ayuda a tomar en cuenta el manto vegetal y la rugosidad de la superficie. La ventaja de esta técnica es su capacidad de reducir o incluso eliminar los efectos de la temperatura de superficie y, por lo tanto, no requiere datos de temperatura auxiliares.

**Normalized Difference Polarization Index (NDPI)**
http://earth.esa.int/landtraining07/Wenetal_TMI_TsMV_JGR_2003.pdf

## 5. Exploración numérica de los datos
El objetivo de esta sección es familiarizarse con la generación de recortes de una escena completa, obtención de histogramas, diagramas de dispersion y estadisticos a partir de las imágenes correspondientes a cada canal del radiometro GMI de GPM.

### 5.1 Obtención de un recorte de la escena completa
Ahora vamos a generar un recorte (subset) de la escena para analizar con detalle y numéricamente la respuesta de las diferentes superficies en cada una de las bandas, mediante diagramas de dispersión e histogramas.

```python
# Definimos los limites del area a recortar:
latmin=-44
latmax=-20
lonmin=-75
lonmax=-45

# Generamos el recorte para cada una de las bandas o canales:
idx1 = (S1_lat>=latmin) & (S1_lat<=latmax) & (S1_lon>=lonmin) & (S1_lon<=lonmax)
idx2 = (S2_lat>=latmin) & (S2_lat<=latmax) & (S2_lon>=lonmin) & (S2_lon<=lonmax)

S1_sub_lat = S1_lat[:,:][idx1]
S1_sub_lon = S1_lon[:,:][idx1]

S2_sub_lat = S2_lat[:,:][idx2]
S2_sub_lon = S2_lon[:,:][idx2]

S1_subch1 = S1_tb[:,:,0][idx1]
S1_subch2 = S1_tb[:,:,1][idx1]
S1_subch3 = S1_tb[:,:,2][idx1]
S1_subch4 = S1_tb[:,:,3][idx1]
S1_subch5 = S1_tb[:,:,4][idx1]
S1_subch6 = S1_tb[:,:,5][idx1]
S1_subch7 = S1_tb[:,:,6][idx1]
S1_subch8 = S1_tb[:,:,7][idx1]
S1_subch9 = S1_tb[:,:,8][idx1]

S2_subch1 = S2_tb[:,:,0][idx2]
S2_subch2 = S2_tb[:,:,1][idx2]
S2_subch3 = S2_tb[:,:,2][idx2]
S2_subch4 = S2_tb[:,:,3][idx2]
```

### 5.4 Clasificación de la imagen a partir de análisis de histogramas
La idea es identificar diferentes coberturas mediante una clasificacion basada en umbrales que podemos extraer del análisis de los histogramas. Es decir, poder generar por ejemplo una imagen que contemple dos categorias: agua y suelo.

```python
# Extraigo las coordenadas de los pixeles que cumplen una condicion, por ejemplo, que la temperatura de brillo en 
# uno de los canales es superior, inferior o esta comprendida entre dos valores. Por ejemplo, observando el
# histograma correspondiente al canal de 10.65V tomo un valor umbral de 200K para separar ambos maximos en la
# distribucion.

tbthr1=200.0
tbthr2=300.0

# Defino un primer conjunto de puntos donde la TB en 10.65V es menor que 200K
latreg1 = S1_sub_lat[np.where(S1_subch1<=tbthr1)]
lonreg1 = S1_sub_lon[np.where(S1_subch1<=tbthr1)]

# Defino un segundo conjunto de puntos donde la TB en 10.65V es mayor que 200K y menor que 300K
latreg2 = S1_sub_lat[np.where((S1_subch1>tbthr1) & (S1_subch1<=tbthr2))]
lonreg2 = S1_sub_lon[np.where((S1_subch1>tbthr1) & (S1_subch1<=tbthr2))]

# Generamos la imagen clasificada, graficando de fondo el campo de Tb de 10.65V y sobre ella con cierto grado de
# transparencia las dos categorias en las cuales separamos nuestros datos.

# Definimos los limites del area a visualizar
latn=-20
lats=-44
lonw=-75
lone=-45

fig = plt.figure(figsize=(20, 10))

ax1 = fig.add_subplot(111,aspect=1)

matplotlib.rc('xtick', labelsize=14)
matplotlib.rc('ytick', labelsize=14)

plt.scatter(lonreg1, latreg1, color='b', marker='o', alpha=0.1, label='Agua')
plt.scatter(lonreg2, latreg2, color='r', marker='o', alpha=0.1, label='Suelo')
plt.legend(loc='upper right', fontsize=16)

plt.plot(mapa['samerica'][:,0],mapa['samerica'][:,1], color='0.0', linewidth=1.0)

plt.title('Mapa clasificado en base al canal 10.65V', fontsize=20)

plt.axis([lonw, lone, lats, latn])

#plt.savefig('ejemplo.png')

plt.show()
```

## 6. Índice diferencial normalizado de polarización (NDPI)

```python
# Cálculo del NDPI para la frecuencia de 10.65 GHz
NDPI1=(S1_tb[:,:,0]-S1_tb[:,:,1])/(0.5*(S1_tb[:,:,0]+S1_tb[:,:,1]))

# Cálculo del NDPI para la frecuencia de 18.7 GHz
NDPI2=(S1_tb[:,:,2]-S1_tb[:,:,3])/(0.5*(S1_tb[:,:,2]+S1_tb[:,:,3]))
```
